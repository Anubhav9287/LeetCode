```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
a = TreeNode(5)
b = TreeNode(3)
c = TreeNode(6)
a.left = b
a.right = c
e = TreeNode(2)
f = TreeNode(4)
b.left = e
b.right = f
g = TreeNode(7)
c.right = g
```

```python
def printT(root):
    if not root:
        return
    print(root.val)
    printT(root.left)
    printT(root.right)
printT(a)
```

```python
key = 3
def deleteNode(root,key):
    if not root:
        return root
    
    if key > root.val:
        root.right = deleteNode(root.right, key)
    elif key < root.val:
        root.left = deleteNode(root.left, key)
    else:
        if not root.left:
            return root.right
        elif not root.right:
            return root.left
        
        #find the min from right subtree
        cur = root.right
        while cur.left:
            cur = cur.left
        root.val = cur.val
        root.right = deleteNode(root.right, root.val)
    return root
printT(deleteNode(a,key))
```
