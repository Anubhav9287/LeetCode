Given the head of a singly linked list, reverse the list, and return the reversed list.


```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
a = ListNode(1)
b = ListNode(2)
c = ListNode(3)
d = ListNode(4)
e = ListNode(5)
a.next = b
b.next = c
c.next = d
d.next = e
```

```python
def printF(head):
    curr = head
    while curr:
        print(curr.val, "--> ", end="")
        curr = curr.next
    print("None")
printF(a)
```

```python
def reverseList(head):
#     temp = ListNode()
    prev = None
    curr = head
    while curr:
        temp = curr.next
        curr.next = prev
        prev = curr
        curr = temp
    return prev
```

```python
printF(reverseList(a))
```
