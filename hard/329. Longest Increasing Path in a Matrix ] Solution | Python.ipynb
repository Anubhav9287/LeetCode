Given an m x n integers matrix, return the length of the longest increasing path in matrix.
<br>
From each cell, you can either move in four directions: left, right, up, or down. You may not move diagonally or move outside the boundary (i.e., wrap-around is not allowed).
<br>


```python
matrix = [[9,9,4],[6,6,8],[2,1,1]]
```

```python
class Solution:
    def longestIncreasingPath(self, matrix) -> int:
        self.memo = {}
        res = 1
        self.directions = [(0,1),(1,0),(-1,0),(0,-1)]
        
        for r in range(len(matrix)):
            for c in range(len(matrix[0])):
                res = max(res,self.dfs(matrix,r,c))
        return res

    def dfs(self,matrix,r,c):
            if (r,c) in self.memo:
                return self.memo[(r,c)]
            self.memo[(r,c)] = 1
            
            #check in all directions
            for d_r,d_c in self.directions:
                new_r = d_r+r
                new_c = d_c+c
                
                #boundry conditions:
                if (0 <= new_r < len(matrix)) and (0 <= new_c < len(matrix[0])) and matrix[r][c] < matrix[new_r][new_c]:
                    self.memo[(r,c)] = max(self.memo[(r,c)], 1+ self.dfs(matrix,new_r,new_c))
            return self.memo[(r,c)]
```

```python
a = Solution()
a.longestIncreasingPath(matrix)
```
