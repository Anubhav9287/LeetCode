You have n binary tree nodes numbered from 0 to n - 1 where node i has two children leftChild[i] and rightChild[i], return true if and only if all the given nodes form exactly one valid binary tree.
<br>
If node i has no left child then leftChild[i] will equal -1, similarly for the right child.
<br>
Note that the nodes have no values and that we only use the node numbers in this problem.<br>
 

```python
# n = 4
# leftChild = [1,-1,3,-1]
# rightChild = [2,3,-1,-1]
n = 4 
leftChild = [1,-1,3,-1]
rightChild = [2,-1,-1,-1]
import collections
def validateBinaryTreeNodes(n, leftChild, rightChild):
    # child can be 0, 1 or 2 || only one parent
    # cycle
    # root should be one --> reach to every other node
    
    #create a graph/tree of given data
    parent_to_child = collections.defaultdict(list)
    child_to_parent = {}
    
    for i in range(n):
        left_child = leftChild[i]
        right_child = rightChild[i]
        
        if left_child != -1:
            parent_to_child[i].append(left_child)
            
            if left_child in child_to_parent:
                return False
            else:
                child_to_parent[left_child] = i
        if right_child != -1:
            parent_to_child[i].append(right_child)
            
            if right_child in child_to_parent:
                return False
            else:
                child_to_parent[right_child] = i
    
    #check root
    root_candiate = None
    
    for i in range(n):
        if i not in child_to_parent:
            if not root_candiate:
                root_candiate = i
            else:
                return False
    
    if root_candiate is None:
        return False
    
    queue = [root_candiate]
    visited = set()
    
    while queue:
        curr = queue.pop(0)
        visited.add(curr)
        for child in parent_to_child[curr]:
            if child not in visited:
                queue.append(child)
    
    return len(visited) == n
                
```

```python
validateBinaryTreeNodes(n, leftChild, rightChild)
```
