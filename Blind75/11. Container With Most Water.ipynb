```python
height = [1,8,6,2,5,4,8,3,7]
```

```python
def maxArea(height):
    l = 0
    r = len(height) - 1
    max_area = 0
    while l < r:
        temp_area = min(height[l],height[r]) * (r-l)
        max_area = max(max_area,temp_area)
        if height[l] < height[r]:
            l += 1
        else:
            r -= 1
    return max_area
```

```python
maxArea(height)
```
