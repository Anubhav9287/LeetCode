```python
class Node:
    def __init__(self,key):
        self.val = key
        self.next = None
#create Nodes:
a = Node(3)
b = Node(2)
c = Node(0)
d = Node(-4)
#Node Connection
a.next = b
b.next = c
c.next = d
d.next = b
```

```python
def checkLinkedListCycle(head):
    mapDist = set()
    while head:
        if head in mapDist:
            return True
        mapDist.add(head)
        head = head.next
    return False
```

```python
checkLinkedListCycle(a)
```

```python
#slow fast pointer method
def checkLinkedlistcycle(head):
    slow,fast = head,head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if fast == slow:
            return True
    return False
```

```python
checkLinkedlistcycle(a)
```
