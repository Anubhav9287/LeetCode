For two strings s and t, we say "t divides s" if and only if s = t + ... + t (i.e., t is concatenated with itself one or more times).
<br>
Given two strings str1 and str2, return the largest string x such that x divides both str1 and str2.
<br>


```python
#TC1
str1 = "ABCABC"
str2 = "ABC"
r1 = "ABC"
#TC2
str3 = "ABABAB"
str4 = "ABAB"
r2 = "AB"
#TC3
str5 = "LEET"
str6 = "CODE"
r3 = ""
#combine
TCs = {
    1 :[str1,str2,r1],
    2 :[str3,str4,r2],
    3 :[str5,str6,r3]
}
```

```python
def gcdOfStrings(str1,str2):
    l1 = len(str1)
    l2 = len(str2)
    
    def isDivisor(l):
        if l1%l != 0 or l2%l != 0:
            return False
        f1,f2 = l1//l,l2//l
        return str1[:l] * f1 == str1 and str1[:l] * f2 == str2 
        
        
    for l in range(min(l1,l2), 0, -1):
        if isDivisor(l):
            return str1[:l]
    return ""
```

```python
gcdOfStrings(str1,str2)
```

```python
#run all test cases
for i in TCs:
    if gcdOfStrings(TCs[i][0],TCs[i][1]) == TCs[i][2]:
        print(i," Test case passed")
    else:
        print(i," Test Case FAILED")
```
