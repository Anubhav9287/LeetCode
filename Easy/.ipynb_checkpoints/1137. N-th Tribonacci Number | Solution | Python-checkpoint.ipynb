The Tribonacci sequence Tn is defined as follows: 
<br>
T0 = 0, T1 = 1, T2 = 1, and Tn+3 = Tn + Tn+1 + Tn+2 for n >= 0.
<br>
Given n, return the value of Tn.
<br>



```python
def tribonacci(n):
    if n == 1:
        return 0
    elif n == 2:
        return 1
    elif n == 3:
        return 1
    p0 = 0
    p1 = 1
    p2 = 1
    p = p0+p1+p2
    for i in range(3,n+1):
        p = p0+p1+p2
        p0 = p1
        p1 = p2
        p2 = p
        
    return p
```

```python
tribonacci(4)
```

```python
def tribonacci2(n):
    t = [0,1,1]
    
    if n < 3:
        return t[n]
    for i in range(3,n+1):
        t[0] , t[1], t[2] = t[1], t[2], sum(t)
    return t[2]
```

```python
tribonacci2(4)
```
