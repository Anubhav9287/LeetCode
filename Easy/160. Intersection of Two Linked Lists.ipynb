```python
class Node:
    def __init__(self,key):
        self.val = key
        self.next = None
#Create A Nodes
a_1 = Node(4)
a_2 = Node(1)
b_1 = Node(5)
b_2 = Node(6)
b_3 = Node(1)
c_1 = Node(8)
c_2 = Node(4)
c_3 = Node(5)
#Connections of Nodes
a_1.next = a_2
a_2.next = c_1
#series B
b_1.next = b_2
b_2.next = b_3
b_3.next = c_1
#Connection list
c_1.next = c_2
c_2.next = c_3
```

```python
#printing the 2 lists
def printList(headNode):
    while headNode:
        print(headNode.val,' --> ',end=' ')
        headNode = headNode.next
    print(' None')
printList(a_1)
printList(b_1)
```

```python
def intersectionNode(head1,head2):
    dic = []
    while head1:
        dic.append(head1)
        head1 = head1.next
    while head2:
        if head2 in dic:
            return head2
        head2 = head2.next
    return None
```
