```python
class Node:
    def __init__(self,key=0,left=None,right=None):
        self.val = key
        self.left = left
        self.right = right
#Create Nodes:
a1 = Node(1)
a2 = Node(2)
a3 = Node(3)
#Connection:
a1.left = a2
a1.right = a3
# New Tree
b1 = Node(1)
b2 = Node(2)
b3 = Node(3)
#Connection:
b1.left = b2
b1.right = b3
```

```python
def sameTree(root1,root2):
    if not root1 and not root2:
        return True
    if not root1 or not root2:
        return False
    if root1.val != root2.val:
        return False
    return sameTree(root1.left,root2.left) and sameTree(root1.right,root2.right)
```

```python
sameTree(a1,b1)
```
