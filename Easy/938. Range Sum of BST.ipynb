```python
class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
#define Nodes
n1 = Node(10)
n2 = Node(5)
n3 = Node(15)
n4 = Node(3)
n5 = Node(7)
n6 = Node(18)
#Connecting the node as per tree
n1.left = n2
n1.right = n3
n2.left = n4
n2.right = n5
n3.right = n6
```

```python
low = 7
high = 15
stack = []
stack.append(n1)
result = 0
while stack:
    current = stack.pop()
    if current.val >= low and current.val<= high:
        result += current.val
#     print(current.val)
    if current.right != None and current.val < high:
        stack.append(current.right)
    if current.left != None and current.val > low:
        stack.append(current.left)
print(result)
```
