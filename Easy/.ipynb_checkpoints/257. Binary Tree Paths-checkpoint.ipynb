```python
class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
#Create Nodes
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(5)
#Connection
n1.left = n2
n1.right = n3
n2.right = n4
```

```python
#firstfunction
def binaryTreePaths(treeNode):
    resultPaths = []
    if treeNode == None:
        return resultPaths
    dfs(treeNode,'',resultPaths)
    return resultPaths
```

```python
def dfs(treeNode, path, resultPaths):
    path += str(treeNode.val)
    if treeNode.left is None and treeNode.right is None:
        resultPaths.append(path)
        return
    if treeNode.left is not None:
        dfs(treeNode.left, path + '-->',resultPaths)
    if treeNode.right is not None:
        dfs(treeNode.right, path+'-->',resultPaths)
```

```python
binaryTreePaths(n1)
```

```python
#Without using recursion
def binaryTreePathWithoutRecursion(treeNode):
    result = []
    stack = [[treeNode,'']]
    while stack:
        current = stack.pop()
        if current[0].left is None and current[0].right is None:
            result.append(current[1]+str(current[0].val))
        if current[0].left is not None:
            stack.append([current[0].left,current[1]+str(current[0].val)+'->'])
        if current[0].right is not None:
            stack.append([current[0].right,current[1]+str(current[0].val)+'->'])
    return result
```

```python
binaryTreePathWithoutRecursion(n1)
```
