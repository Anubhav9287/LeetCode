```python
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
```

```python
a = Node(7)
b = Node(13)
c = Node(11)
d = Node(10)
e = Node(1)
#Connection
a.next = b
b.next = c
c.next = d
d.next = e
#random Connection
a.random = e
b.random = a
c.random = e
d.random = c
e.random = a
```

```python
def printRandom(root):
    curr = root
    while curr:
        print(curr.val , " --> Random Node --> ",curr.random.val)
        curr = curr.next
```

```python
printRandom(a)
```

```python
def copyRandomList(head):
    copyToMap = {None:None}
    
    curr = head
    while curr:
        copy = Node(curr.val)
        copyToMap[curr] = copy
        curr = curr.next
    
    curr = head
    
    while curr:
        copy = copyToMap[curr]
        copy.next = copyToMap[curr.next]
        copy.random = copyToMap[curr.random]
        curr = curr.next
    return copyToMap[head]
```

```python
copyRandomList(a)
```

```python
printRandom(copyRandomList(a))
```
