Given an array of strings strs, group the anagrams together. You can return the answer in any order.
<br>
An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.<br>


```python
strs = ["eat","tea","tan","ate","nat","bat"]
```

```python
def groupAnagrams(strs):
    
    word_map = {}
    result = []
    for i in strs:
        temp = ''.join(sorted(i))
        if temp in word_map:
            word_map[temp].append(i)
        else:
            word_map[temp] = [i]
    
    for each in word_map:
        result.append(word_map[each])
    return result
```

```python
groupAnagrams(strs)
```
