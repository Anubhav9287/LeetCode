```python
class Node:
    def __init__(self,key):
        self.val = key
        self.next = None
#create Nodes:
a = Node(1)
b = Node(2)
c = Node(3)
d = Node(4)
e = Node(5)
```

```python
def createFreshConnection(a,b,c,d,e):
    #Connection:
    a.next = b
    b.next = c
    c.next = d
    d.next = e
    e.next = None
```

```python
createFreshConnection(a,b,c,d,e)
```

```python
def printList(head):
    while head:
        print(head.val,' --> ',end='')
        head = head.next
    print('None')
```

```python
printList(a)
```

```python
#Remove 2nd element
def remove_n_element(head,n):
    tmp = head
    for i in range(n-2):
        tmp = tmp.next
    tmp.next = tmp.next.next
```

```python
remove_n_element(a,2)
printList(a)
```

```python
createFreshConnection(a,b,c,d,e)
printList(a)
```

```python
#reverse the linkedlist
def reverseLinkedList(head):
    prev = None
    current = head
    while current:
        nxt = current.next
        current.next = prev
        prev = current
        current = nxt
    return prev
```

```python
printList(reverseLinkedList(a))
```

```python
#normal list --> reverse --> remove 2nd element --> again reverse
createFreshConnection(a,b,c,d,e)
printList(a)
```

```python
#normal list --> reverse --> remove 2nd element --> again reverse
x = reverseLinkedList(a)
remove_n_element(x,2)
printList(reverseLinkedList(x))
```

```python
createFreshConnection(a,b,c,d,e)
printList(a)
```

```python
n = 2
dummy = Node(0)
dummy.next = a
left = dummy
right = a
while n > 0 and right:
    right = right.next
    n -= 1

while right:
    right = right.next
    left = left.next

left.next = left.next.next
printList(dummy.next)
```
