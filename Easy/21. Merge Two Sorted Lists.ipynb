You are given the heads of two sorted linked lists list1 and list2. <br>

Merge the two lists in a one sorted list. The list should be made by splicing together the nodes of the first two lists.<br>

Return the head of the merged linked list.<br>


```python
class Node:
    def __init__(self,key):
        self.val = key
        self.next = None
#Create Nodes:
a1 = Node(1)
a2 = Node(2)
a3 = Node(4)
b1 = Node(1)
b2 = Node(3)
b3 = Node(4)
#Connection
a1.next = a2
a2.next = a3
b1.next = b2
b2.next = b3
```

```python
def printList(head):
    while head:
        print(head.val,' --> ',end='')
        head = head.next
    print('None')
```

```python
printList(a1)
printList(b1)
```

```python
#Just printing the merged list
def mergeSortedList(head1,head2):
    while head1 and head2:
        if head1.val < head2.val:
            print(head1.val,' --> ',end='')
            head1 = head1.next
        else:
            print(head2.val,' --> ',end='')
            head2 = head2.next
    while head1:
        print(head1.val,' --> ',end='')
        head1 = head1.next
    while head2:
        print(head2.val,' --> ',end='')
        head2 = head2.next
    print('None')
```

```python
mergeSortedList(a1,b1)
```

```python
def mergeSortedLists(head1,head2):
    dummy = Node(0)
    tail = dummy
    while head1 and head2:
        if head1.val < head2.val:
            tail.next = head1
            head1 = head1.next
        else:
            tail.next = head2
            head2 = head2.next
        tail = tail.next
    if head1:
        tail.next = head1
    if head2:
        tail.next = head2
    return dummy.next
```

```python
ab = mergeSortedLists(a1,b1)
```

```python
printList(ab)
```
