## Sample 1:
Input: root = [3,4,5,1,2], subRoot = [4,1,2] <br>
Output: true <br>


```python
class TreeNode:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
```

```python
#First Tree:
a = TreeNode(3)
a1 = TreeNode(4)
a2 = TreeNode(5)
a3 = TreeNode(1)
a4 = TreeNode(2)
a.left = a1
a.right = a2
a1.left = a3
a1.right = a4
```

```python
#Subtree
s = TreeNode(4)
s1 = TreeNode(1)
s2 = TreeNode(2)
s.left = s1
s.right = s2
```

```python
def isSameTree(r,s):
    if r == None or s == None:
        return s==None and r == None
    elif r.val == s.val:
        return isSameTree(r.left,s.left) and isSameTree(r.right,s.right)
    else:
        return False
```

```python
def isSubtree(root,subRoot):
    if root == None:
        return False
    elif isSameTree(root, subRoot):
        return True
    else:
        return isSubtree(root.left, subRoot) or isSubTree(root.right, subRoot)
```

```python
isSubtree(a,s)
```
