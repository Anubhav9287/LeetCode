From any string, we can form a subsequence of that string by deleting some number of characters (possibly no deletions).
<br>
Given two strings source and target, return the minimum number of subsequences of source such that their concatenation equals target. If the task is impossible, return -1.<br>


```python
source = "abc"
target = "abcbc"
```

```python
def shortestWay(source,target):
    target_idx = res = 0
    
    while target_idx < len(target):
        prev_target_idx = target_idx
        
        for char in source:
            if target_idx < len(target) and char == target[target_idx]:
                target_idx += 1
        
        if prev_target_idx == target_idx:
            return -1
        res += 1
    return res
```

```python
shortestWay(source,target)
```
