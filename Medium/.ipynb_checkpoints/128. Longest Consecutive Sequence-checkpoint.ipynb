Given an unsorted array of integers nums, return the length of the longest consecutive elements sequence.<br>
You must write an algorithm that runs in O(n) time.<br>
<br>


```python
nums = [100,4,200,1,3,2]
```

```python
nums = set(nums)
```

```python
nums
```

```python
l = 0
m = 0
for i in nums:
    v = i
    if v-1 not in nums:
#         print('Start: ',v)
        while v in nums:
#             print(v)
            l += 1
            v = v + 1
        m = max(m,l)
        l = 0
```
