```python
tokens = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
```

```python
ops = ['+','-','*','/']
```

```python
stack = []
for i in tokens:
    if i not in ops:
        stack.append(i)
    else:
        if i == '+':
            val1 = stack.pop()
            val2 = stack.pop()
            stack.append(int(val1)+int(val2))
        elif i == '*':
            val1 = stack.pop()
            val2 = stack.pop()
            stack.append(int(val1)*int(val2))
        elif i == '/':
            val1 = stack.pop()
            val2 = stack.pop()
            stack.append(int(val2)/int(val1))
        elif i == '-':
            val1 = stack.pop()
            val2 = stack.pop()
            stack.append(int(val2)-int(val1))
```
