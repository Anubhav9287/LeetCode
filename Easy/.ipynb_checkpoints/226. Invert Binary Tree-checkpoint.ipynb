```python
class TreeNode:
    def __init__(self, key, left=None, right=None):
        self.val = key
        self.left = left
        self.right = right
#create nodes:
a = TreeNode(4)
b = TreeNode(2)
c = TreeNode(7)
d = TreeNode(1)
e = TreeNode(3)
f = TreeNode(6)
g = TreeNode(9)
#Connection of nodes in form of tree
a.left = b
a.right = c
b.left = d
b.right = e
c.left = f
c.right = g
```

```python
#breadth first search
def printTree(root):
    queue = []
    result = []
    queue.append(root)
    while queue:
        current = queue.pop(-1)
        result.append(current.val)
        if current.right:
            queue.append(current.right)
        if current.left:
            queue.append(current.left)
    print(result)
```

```python
printTree(a)
```

```python
def invertTree(root):
    if not root:
        return None
    tmp  = root.left
    root.left = root.right
    root.right = tmp
    invertTree(root.left)
    invertTree(root.right)
```

```python
invertTree(a)
```

```python
printTree(a)
```
