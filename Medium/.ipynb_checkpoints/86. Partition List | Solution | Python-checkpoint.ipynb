Given the head of a linked list and a value x, partition it such that all nodes less than x come before nodes greater than or equal to x.
<br>
You should preserve the original relative order of the nodes in each of the two partitions.
<br>


```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
a = ListNode(1)
b = ListNode(4)
c = ListNode(3)
d = ListNode(2)
e = ListNode(5)
f = ListNode(2)
a.next = b
b.next = c
c.next = d
d.next = e
e.next = f
```

```python
def printF(root):
    curr = root
    while curr:
        print(curr.val," --> ",end='')
        curr = curr.next
    print("None")
printF(a)
```

```python
def partition(head, x):
    l1 = p1 = ListNode(0)
    l2 = p2 = ListNode(0)
    curr = head
    
    while curr:
        if curr.val >= x:
            p2.next = curr
            p2 = p2.next
        else:
            p1.next = curr
            p1 = p1.next
        curr = curr.next
    p2.next = None
    p1.next = l2.next
    return l1.next
```

```python
x = 3
printF(partition(a, x))
```
