```python
class Node:
    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key
#Create Nodes
a = Node(1)
b = Node(3)
c = Node(2)
d = Node(4)
e = Node(5)
#Node Connection
a.right = b
b.left = c
b.right = d
d.right = e
```

```python
def findLongestConsecutiveSequence(root, count, target, max_val):
    if root is None:
        return
    elif root.val == target:
        count += 1
    else:
        count = 1
    max_val[0]  = max(count,max_val[0])
    findLongestConsecutiveSequence(root.left, count, root.val + 1, max_val)
    findLongestConsecutiveSequence(root.right, count, root.val + 1, max_val)
```

```python
max_val = [0]
findLongestConsecutiveSequence(a, 0,0,max_val)
print(max_val[0])
```
