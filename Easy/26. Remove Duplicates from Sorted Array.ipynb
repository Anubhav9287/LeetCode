```python
def RemoveDuplicate(nums):
    l = 1
    r = 1
    if len(nums) <= 1:
        return len(nums)
    for i in range(1,len(nums)):
        if nums[i] == nums[i-1]:
            r += 1
        else:
            nums[l] = nums[i]
            l += 1
            r += 1
    return l
```
