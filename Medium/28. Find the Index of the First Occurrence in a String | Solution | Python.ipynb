```python
haystack = "sabutsads"
needle = "sad"
```

```python
def strStr(haystack,needle):
    for i in range(len(haystack)-len(needle)+1):
        if haystack[i] == needle[0]:
            if haystack[i:i+len(needle)] == needle:
                return i
    return -1
    
strStr(haystack,needle)
```
