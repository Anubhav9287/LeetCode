```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
a = TreeNode(3)
b = TreeNode(9)
c = TreeNode(20)
d = TreeNode(15)
e = TreeNode(7)
a.left = b
a.right = c
c.left = d
c.right = e
```

```python
def printf(root):
    if not root:
        return None
    print(root.val, end=' >> ')
    if root.left:
        printf(root.left)
    if root.right:
        printf(root.right)
```

```python
printf(a)
```

```python
def maxDepth(root):
    stack = []
    if root:
        stack.append((root,1))
    height = 0
    while stack:
        curr,h = stack.pop()
        height = max(height,h)
        if curr.left:
            stack.append((curr.left,h+1))
        if curr.right:
            stack.append((curr.right,h+1))
    return height
```
