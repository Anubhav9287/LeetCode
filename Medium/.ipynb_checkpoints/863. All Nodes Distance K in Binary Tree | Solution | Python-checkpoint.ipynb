Given the root of a binary tree, the value of a target node target, and an integer k, return an array of the values of all nodes that have a distance k from the target node.
<br>
You can return the answer in any order.
<br>


```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
a = TreeNode(3)
b = TreeNode(5)
c = TreeNode(1)
a.left = b
a.right = c
d = TreeNode(6)
e = TreeNode(2)
f = TreeNode(0)
g = TreeNode(8)
b.left = d
b.right = e
c.left = f
c.right = g
h = TreeNode(7)
i = TreeNode(4)
e.left = h
e.right = i
```

```python
# 3 --> 5,1
# 5 --> 6,2,3
# 1 --> 0,8,3
# 2 --> 7,4,5
# 6 --> 5
# 0 --> 1
# 8 --> 1
# 7 --> 2
# 4 -- > 2
import collections
def create_graph(node, parent, graph):
    if not node:
        return
    if parent:
        graph[node.val].append(parent.val)
    if node.left:
        graph[node.val].append(node.left.val)
    if node.right:
        graph[node.val].append(node.right.val)
    create_graph(node.left, node, graph)
    create_graph(node.right, node, graph)

def search(graph,target,k):
    queue = [target]
    visited = set()
    while queue and k > 0:
        for _ in range(len(queue)):
            cur = queue.pop(0)
            visited.add(cur)
            for n in graph[cur]:
                if n not in visited:
                    queue.append(n)
        k -= 1
    return queue
    
def distanceK(root, target, k):
    graph = collections.defaultdict(list)
    
    create_graph(root, None ,graph)
    
    print(search(graph,target,k))
```

```python
distanceK(a,5,2)
```
