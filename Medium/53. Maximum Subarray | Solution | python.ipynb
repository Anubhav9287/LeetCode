Given an integer array nums, find the subarray with the largest sum, and return its sum.
<br>


```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
```

```python
def maxSubArray(nums):
    temp = [0] * len(nums)
    temp[0] = nums[0]
    m = nums[0]
    for i in range(1,len(nums)):
        temp[i] = max(nums[i], temp[i-1]+nums[i])
        m = max(m,temp[i])
    return m
```

```python
maxSubArray(nums)
```
