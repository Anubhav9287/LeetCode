```python
class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None
#create Nodes:
a1 = Node(6)
a2 = Node(2)
a3 = Node(8)
a4 = Node(0)
a5 = Node(4)
a6 = Node(3)
a7 = Node(5)
a8 = Node(7)
a9 = Node(9)
#Node Connection:
a1.left = a2
a1.right = a3
a2.left = a4
a2.right = a5
a5.left = a6
a5.right = a7
a3.left = a8
a3.right = a9
```

```python
def printDfs(root):
    result = []
    stack = []
    stack.append(root)
    while stack:
        cur = stack.pop()
        result.append(cur.val)
        if cur.right:
            stack.append(cur.right)
        if cur.left:
            stack.append(cur.left)
    return result
```

```python
printDfs(a1)
```

```python
def lowestCommonAncestor(root,p,q):
    cur = root
    while cur:
        if p.val > cur.val and q.val > cur.val:
            cur = cur.right
        elif p.val < cur.val and q.val < cur.val:
            cur = cur.left
        else:
            return cur.val
```

```python
lowestCommonAncestor(a1,a2,a3)
```
