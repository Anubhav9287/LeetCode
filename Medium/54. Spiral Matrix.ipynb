Given an m x n matrix, return all elements of the matrix in spiral order.
<br>

<br>Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]
<br>Output: [1,2,3,6,9,8,7,4,5]

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
```

```python
def SpiralMatrix(matrix):
    result = []
    t = 0
    r = len(matrix[0])
    d = len(matrix)
    l = 0
    while l < r and t < d:
        for i in range(l,r):
            result.append(matrix[t][i])
        t = t + 1
        for i in range(t,d):
            result.append(matrix[i][r-1])
        r -= 1
        if not (l<r and t<d):
            break
        for i in range(r-1,l-1,-1):
            result.append(matrix[d-1][i])
        d -= 1
        for i in range(d-1,t-1,-1):
            result.append(matrix[i][l])
        l += 1
    return result
```

```python
SpiralMatrix(matrix)
```
