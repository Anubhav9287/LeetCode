```python
class Node:
    def __init__(self,key):
        self.val = key
        self.next = None
#Create Nodes
a = Node(1)
b = Node(2)
c = Node(3)
d = Node(4)
e = Node(5)
#Node connection
a.next = b
b.next = c
c.next = d
d.next = e
```

```python
def printLinkedList(head):
    while head:
        print(head.val,' --> ',end='')
        head = head.next
    print('None')
```

```python
printLinkedList(a)
```

```python
def reverseList(head):
    prev = None
    current = head
    while current:
        nxt = current.next
        current.next = prev
        prev = current
        current = nxt
    return prev
```

```python
output = reverseList(a)
```

```python
printLinkedList(output)
```
