```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
a = TreeNode(1)
b = TreeNode(2)
c = TreeNode(3)
a.left = b
a.right = c
a1 = TreeNode(1)
b1 = TreeNode(2)
a2 = TreeNode(1)
b2 = TreeNode(2)
a1.left = b1
a2.right = b2
```

```python
def isSameTree(p, q):
    if not p and not q:
        return True
    if p and q and q.val == p.val:
        return isSameTree(p.left,q.left) and isSameTree(p.right,q.right)
    else:
        return False
```

```python
isSameTree(a,a)
```
