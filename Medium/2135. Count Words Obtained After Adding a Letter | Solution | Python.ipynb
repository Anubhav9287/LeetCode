You are given two 0-indexed arrays of strings startWords and targetWords. Each string consists of lowercase English letters only.
<br>
For each string in targetWords, check if it is possible to choose a string from startWords and perform a conversion operation on it to be equal to that from targetWords.
<br>
The conversion operation is described in the following two steps:
<br>
Append any lowercase letter that is not present in the string to its end.<br>
For example, if the string is "abc", the letters 'd', 'e', or 'y' can be added to it, but not 'a'. If 'd' is added, the resulting string will be "abcd".<br>
Rearrange the letters of the new string in any arbitrary order.<br>
For example, "abcd" can be rearranged to "acbd", "bacd", "cbda", and so on. Note that it can also be rearranged to "abcd" itself.<br>
Return the number of strings in targetWords that can be obtained by performing the operations on any string of startWords.<br>

Note that you will only be verifying if the string in targetWords can be obtained from a string in startWords by performing the operations. The strings in startWords do not actually change during this process.<br>


```python
startWords = ["ant","act","tack"]
targetWords = ["tack","act","acti"]
```

```python
def wordCount(startWords, targetWords):
    
    start_words = set()
    res = 0
    for word in startWords:
        start_words.add("".join(sorted(word)))
    
    for word in targetWords:
        word = "".join(sorted(word))
        
        for i in range(len(word)):
            if word[:i]+word[i+1:] in start_words:
                res+=1
                break
    return res
    
```

```python
wordCount(startWords, targetWords)
```
