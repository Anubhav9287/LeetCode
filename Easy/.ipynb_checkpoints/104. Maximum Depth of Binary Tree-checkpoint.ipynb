```python
class TreeNode:
    def __init__(self,key=0,left=None,right=None):
        self.val = key
        self.left = left
        self.right = right
#create Nodes:
a = TreeNode(3)
b = TreeNode(9)
c = TreeNode(20)
d = TreeNode(15)
e = TreeNode(7)
#connection:
a.left = b
a.right = c
c.left = d
c.right = e
```

```python
def printBFS(root):
    queue = []
    queue.append(root)
    result = []
    count = 1
    while queue:
        cur = queue.pop(-1)
        result.append(cur.val)
        if cur.right:
            queue.append(cur.right)
        if cur.left:
            queue.append(cur.left)
        if cur.left or cur.right:
            count += 1
    print(result,' height: ',count)
```

```python
printBFS(a)
```

```python
def maxDepth(root):
    if not root:
        return 0
    return 1 + max(maxDepth(root.left),maxDepth(root.right))
```

```python
maxDepth(a)
```
