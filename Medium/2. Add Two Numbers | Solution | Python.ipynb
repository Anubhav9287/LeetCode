```python
#Create LinkedList
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
#First LinkedList
a = ListNode(2)
a.next = ListNode(4)
a.next.next = ListNode(3)
b = ListNode(5)
b.next = ListNode(6)
b.next.next = ListNode(4)
```

```python
#print LinkedList
def printL(root):
    while root:
        print(root.val,end=' --> ')
        root = root.next
    print("None")
printL(a)
printL(b)
```

```python
def addTwoNumbers(l1,l2):
    carry = 0
    dummyNode = ListNode(0)
    curr = dummyNode
    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0
        new_val = val1+val2+carry
        carry = new_val//10
        curr.next = ListNode(new_val%10)
        curr = curr.next
        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None
    return dummyNode.next
printL(addTwoNumbers(a,b))
```
