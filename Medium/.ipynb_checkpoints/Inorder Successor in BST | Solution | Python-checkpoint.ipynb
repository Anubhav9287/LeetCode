```python
class Node:
    def __init__(self,key):
        self.val = key
        self.left = None
        self.right = None

a = Node(20)
b = Node(9)
c = Node(25)
a.left = b
a.right = c
d = Node(5)
e = Node(12)
b.left = d
b.right = e
f = Node(11)
g = Node(14)
e.left = f
e.right = g
```

```python
# Normal inorder process
ap = []
def printF(root):
    if not root:
        return
    printF(root.left)
    ap.append(root.val)
    printF(root.right)
    return ap
```

```python
printF(a)
```

```python
def Inorder(root,t):
    s = None
    while root:
        if t >= root.val:
            root = root.right
        else:
            s = root
            root = root.left
    return s.val
```

```python
Inorder(a,11)
```
