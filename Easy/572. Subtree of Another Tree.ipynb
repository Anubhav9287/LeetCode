```python
class Node:
    def __init__(self,key=0,left=None,right=None):
        self.val = key
        self.left = left
        self.right = right
#Create Nodes:
a1 = Node(3)
a2 = Node(4)
a3 = Node(5)
a4 = Node(1)
a5 = Node(2)
#connection:
a1.left = a2
a1.right = a3
a2.left = a4
a2.right = a5
#lets say that big tree is node a1 and subtree is a2
b1 = Node(4)
b2 = Node(2)
b3 = Node(1)
b1.left = b2
b1.right = b3
```

```python
def isMatch(root1,root2):
    if not root1 and not root2:
        return True
    if root1.val != root2.val:
        return False
    return isMatch(root1.left,root2.left) and isMatch(root1.right,root2.right)
```

```python
def subTreeOfAnotherTree(root,subtree):
    if isMatch(root,subtree):
        return True
    if not root:
        return False
    return isMatch(root.left,subtree) or isMatch(root.right,subtree)
```

```python
subTreeOfAnotherTree(a1,b1)
```
